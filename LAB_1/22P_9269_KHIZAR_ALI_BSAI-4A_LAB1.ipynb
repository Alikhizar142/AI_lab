# Making Class


```python
class Student:
    def __init__(self,Roll,Name,Marks):
        self.Roll=Roll
        self.Name=Name
        self.Marks=Marks
```

```python
database=[]

def add_student(roll_number,Name,Marks):
    student=Student(roll_number,Name,Marks)
    database.append(student)
    print(f"The student {Name} is successfully Added")

def display_student_details(roll_number):
    for student in database:
        if student.Roll is roll_number:
            print("Details of the Student are as follows")
            print(f"Name : {student.Name}")
            print(f"Roll Number :{student.Roll} ")
            print("Student Marks")
            print(f"Subject Name :  Marks ")
            for sub , mar in student.Marks.items():
                print(f"{sub} : {mar}")
def display_all_students():
    print(" Details of the  all  Students are as follows")
    for student in database:
        print(f"Name : {student.Name}")
        print(f"Roll Number :{student.Roll} ")
        print(f"Student Marks")
        print(f"Subject Name :  Marks ")
        for sub , mar in student.Marks.items():
                print(f"{sub} : {mar}")



def calculate_average_marks(roll_number):
     for student in database:
          if student.Roll is roll_number:
               marks=student.Marks.values()
               avg = sum(marks)/len(marks)
               print(f"The Average marks of student is {avg}")


def find_topper():
     topper=max(database,key=lambda student:sum(student.Marks.values()))
     print(f"Name : {topper.Name}")
     print(f"Roll Number :{topper.Roll} ")
     print(f"Student Marks = {sum(topper.Marks.values())/ len(topper.Marks.values())}")

def main():
     while True:
          print("Menu")
          print("1. Add a new student details")
          print("2. Display details of a specific student")
          print("3. Display details of all students")
          print("4. Calculate and display the average marks of a student")
          print("5. Find the Topper")
          print("6. Exit")
          choice=int(input("Enter your choice"))
          if choice==1:
               roll_number = int(input("Enter Roll Number: "))
               name = input("Enter Name: ")
               marks = {}
               num_subjects = int(input("Enter the number of subjects: "))
               for _ in range(num_subjects):
                    subject = input("Enter Subject: ")
                    mark = float(input(f"Enter Mark for {subject}: "))
                    marks[subject] = mark

               add_student(roll_number,name,marks)


          elif choice==2:
               print("Enter the Roll No of the Student You want")
               roll_number=int(input("Enter Student Roll Number"))
               display_student_details(roll_number)

          elif choice==3:
               display_all_students()
          elif choice==4:
               roll_number = int(input("Enter The Roll Number Of The Student"))
               calculate_average_marks(roll_number)
          elif choice==5:
               find_topper()
          else:
               print("Thanks for using Exiting")
               break




main()
```
