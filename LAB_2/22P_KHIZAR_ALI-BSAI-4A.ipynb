## Task1


```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
df.isnull().sum()
age=df[["age"]]
mean =age.mean()
df.replace(np.nan,mean,inplace=True)
# df.isnull().sum()
connection=df[["connections"]]
mean=connection.mean()
df.replace(np.nan,mean,inplace=True)
df.isnull().sum()

# df.head(100)
```

```python
graph=nx.from_pandas_edgelist(df)

options = {
    "font_size": 10,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "blue",
    "linewidths":4,
    "width": 4,
}
# print(graph.edges)
nx.draw(graph,**options, with_labels=True)
# nx.spring_layout(graph)
```

```python
age=df["age"]
connection=df["connections"]
plt.hist(age,bins=20)
```

```python

# Create data
x = np.linspace(0, 10,10)
y = 2*x + 1
cos_y = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y, label='Straight line', marker='o', linestyle='-')
ax.plot(x, cos_y, label='Cosine', marker='x', linestyle='-')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
plt.show()
```
